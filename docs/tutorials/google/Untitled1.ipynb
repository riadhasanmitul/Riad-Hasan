```python
import matplotlib.pyplot as plt

def DDA(x0, y0, x1, y1):
    dx = abs(x1 - x0)
    dy = abs(y1 - y0)

    Max_Steps = max(dx, dy)

    xincrement = (x1 - x0) / Max_Steps
    yincrement = (y1 - y0) / Max_Steps

    x = float(x0)
    y = float(y0)

    xcoordinates = []
    ycoordinates = []

    for i in range(Max_Steps + 1):
        xcoordinates.append(round(x))
        ycoordinates.append(round(y))

        print(f'Step {i}: x = {round(x)}, y = {round(y)}')

        x += xincrement
        y += yincrement

    plt.plot(xcoordinates, ycoordinates, marker='o', markersize=3, markerfacecolor='red')
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.title("DDA Line Drawing Algorithm")
    plt.grid()
    plt.show()

x0 = 10
y0 = 20
x1 = 50
y1 = 80

DDA(x0, y0, x1, y1)
```
